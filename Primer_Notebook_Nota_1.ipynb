# Problema de Ingeniería (Nivel Básico)

Calcular **el área de un terreno rectangular**.

En ingeniería es común calcular áreas para:

* Diseño de parcelas agrícolas
* Cálculo de cimentaciones
* Planeación de riego

La fórmula es:

[
Área = base \times altura
]

---

# 1. Análisis del Problema

Primero definimos **Entradas, Proceso y Salidas**.

### Entradas

Datos que el usuario debe ingresar:

* base del terreno (metros)
* altura del terreno (metros)

### Proceso

Operación matemática:

* multiplicar base × altura

### Salida

Resultado que mostrará el programa:

* área del terreno en **metros cuadrados (m²)**

---

# 2. Algoritmo (Explicación)

El algoritmo sigue estos pasos:

1. Pedir al usuario la base del terreno.
2. Pedir la altura del terreno.
3. Calcular el área multiplicando base por altura.
4. Mostrar el resultado.

---

# 3. Pseudocódigo

Antes de escribir Python, lo representamos de forma lógica.

```
INICIO

leer base
leer altura

area = base * altura

mostrar area

FIN
```

---

# 4. Implementación en Python

```python
# Programa para calcular el área de un terreno rectangular

# Entrada de datos
base = float(input("Ingrese la base del terreno en metros: "))
altura = float(input("Ingrese la altura del terreno en metros: "))

# Proceso
area = base * altura

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 5. Ejemplo de ejecución

Si el usuario ingresa:

```
base = 20
altura = 15
```

El programa calcula:

```
area = 20 * 15
area = 300
```

Resultado:

```
El área del terreno es: 300 m²
```

---

# 6. Conceptos de Python que usaste

* **Variables:** `base`, `altura`, `area`
* **Operador matemático:** `*` (multiplicación)
* **Entrada de datos:** `input()`
* **Conversión de tipo:** `float()`
* **Salida de datos:** `print()`

### Un ejemplo aplicado a la Ingeniería

Para mostrarte cómo funciona, pensemos en un problema común: **Calcular el volumen de agua necesario para un reservorio cilíndrico** en un proyecto de riego o una obra civil.

En Python, no solo buscamos el resultado, sino la **claridad**. Mira este esquema lógico:

Aquí tienes cómo se vería un fragmento de código siguiendo las **buenas prácticas** que exigiremos en este curso:

````python
import math

def calcular_volumen_reservorio(radio_m, altura_m):
    """
    Calcula el volumen de un reservorio cilíndrico en metros cúbicos.
    """
    area_base = math.pi * (radio_m ** 2)
    volumen_m3 = area_base * altura_m
    return volumen_m3

# Datos de entrada con nombres descriptivos
radio_input = 2.5  # metros
altura_input = 4.0 # metros

# Llamada a la función
resultado = calcular_volumen_reservorio(radio_input, altura_input)

print(f"El volumen del reservorio es: {resultado:.2f} m3")
````
