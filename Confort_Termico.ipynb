## **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere al estado en el que los animales no experimentan estrés por calor o frío, lo que les permite mantener su temperatura corporal central dentro de un rango óptimo con un mínimo gasto de energía. Este estado es crucial para la salud, el bienestar y la productividad de los animales, especialmente en sistemas de producción intensiva.

**Factores que Afectan el Confort Térmico:**

1.  **Temperatura Ambiental:** La temperatura del aire es un factor directo que afecta la transferencia de calor entre el animal y su entorno.
2.  **Humedad Relativa:** Una alta humedad en combinación con altas temperaturas reduce la capacidad del animal para disipar calor por evaporación (sudoración y jadeo).
3.  **Radiación Solar:** La exposición directa al sol aumenta la carga de calor en los animales.
4.  **Velocidad del Viento:** El viento puede ayudar a la disipación de calor por convección en ambientes cálidos, pero puede aumentar la pérdida de calor en ambientes fríos.
5.  **Precipitación:** La lluvia puede enfriar a los animales, pero si se combina con bajas temperaturas y viento, puede causar hipotermia.
6.  **Características del Animal:** La raza, la edad, el peso, el estado fisiológico (lactancia, gestación) y el color del pelaje influyen en la capacidad del animal para tolerar el estrés térmico.

**Mecanismos de Termorregulación:**

Los bovinos tienen varios mecanismos para mantener su temperatura corporal:

*   **Disipación de Calor:**
    *   **Evaporación:** Jadeo (aumento de la frecuencia respiratoria) y sudoración (aunque menos eficiente que en otras especies).
    *   **Convección:** Transferencia de calor al aire circundante.
    *   **Radiación:** Intercambio de calor con objetos cercanos (sol, sombra).
    *   **Conducción:** Transferencia de calor por contacto directo con superficies más frías.
*   **Producción de Calor:** A través del metabolismo de los alimentos y la actividad muscular.

**Índice de Temperatura y Humedad (ITH):**

El ITH es una herramienta ampliamente utilizada para evaluar el riesgo de estrés térmico en el ganado. Combina la temperatura de bulbo seco y la temperatura de bulbo húmedo para dar una medida del impacto combinado del calor y la humedad en los animales. Diferentes rangos de ITH se correlacionan con distintos niveles de estrés térmico, desde

# **Métodos y cálculos**

+ Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72 * (T_{bs} + T_{bh}) + 40.6$$

Donde:

1. $T_{bs}$ Temperatura de bulbo seco ($^oC$)

2. $T_{bh}$ Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$ ambiente no estresante
2. Sí $ITH > 70 \space and \space ITH<= 78$ situación de alerta por estrés térmico
3. Sí $ITH > 78 \space and \space <= 83$ peligro por estrés térmico
4. Sí $ITH > 83$ situación crítica

````python
# ENTRADA: pedir tbs (temperatura de bulbo seco) y tbh (temperatura de bulbo humedo)
tbs = float(input("Ingrese la temperatura de bulto seco, Grados Centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados Centigrados: "))

# PROCESO: Calculo de ith (indice de temperatura y humedad)
ith = 0.72 * (tbs + tbh) + 40.6

# SALIDA: imprimir resultado, de la evaluación del ith y el confort termico
if ith <= 70:
  print ("el ambiente es no estresante")

elif ith > 70 and ith <= 78:
 print ("la situacion de alerta por estrés térmico")

elif ith > 78 and ith <= 83:
  print("la situacion es critica")

else:
  print("la situacion es critica")
````
